# Load Data


```python
# load the dataset skipping the first 3 rows
import pandas as pd
data = pd.read_excel('GDP.xls', skiprows = 3)
data
```

# Clean Data

```python
# remove aggregated data (eg. World Total, Europe Total, Asia Total, etc.)
codes = ['WLD', 'HIC', 'OED', 'PST', 'IBT', 'ECS', 'LMY', 'MIC', 'IBD', 'EUU', 'EAS', 'NAC', 'UMC',
         'LTE', 'EMU', 'EAR', 'EAP', 'TEA', 'LCN', 'TLA', 'LAC', 'LMC', 'TEC', 'ECA', 'MEA', 'ARB',
         'SAS', 'TSA', 'IDA', 'MNA', 'TMN', 'SSF', 'TSS', 'SSA', 'CEB', 'PRE', 'IDX', 'IDB', 'LDC',
         'FCS', 'HPC', 'SST', 'LIC', 'OSS', 'CSS', 'PSS']        
for code in codes:
  data.drop(data.loc[data['Country Code'] == code].index, inplace = True)
data          
```

```python
# after removing some rows, the row index will have missing values, reset to consecutive running order
data = data.reset_index(drop = True)

# drop unnecessary columns
data = data.drop(columns = ['Country Code', 'Indicator Name', 'Indicator Code'], axis = 1)
data
```

```python
# rename column index, so that string '1960' becomes integer 1960
data.columns = ['Name', 1960, 1961, 1962, 1963, 1964, 1965, 1966, 1967, 1968, 1969, 1970, 1971, 1972,
                1973, 1974, 1975, 1976, 1977, 1978, 1979, 1980, 1981, 1982, 1983, 1984, 1985, 1986,
                1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000,
                2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014,
                2015, 2016, 2017, 2018, 2019, 2020]
```

# Create New Data File

```python
# create new df in a desired format
df = pd.DataFrame({ 'Country':[], 'Year':[], 'GDP':[] })
df['Year'] = df['Year'].astype(int)

# start to read data into the new df
r = data.shape[0]       # 220 rows
c = data.shape[1] - 1   # 61 columns (after minus country name)
for row in range(r):
  for i in range(c):    # 61: 1960 to 2020
   df = pd.concat([df, pd.DataFrame({ 'Country':[data['Name'][row]], 'Year':[1960+i], 'GDP':[data[1960+i][row]] }) ])
df
```

```python
# fill null field with zero
df.gdp = df.GDP.fillna(0)
df
```

#Save Final Data

```python
# save final data format for creating Tableau dashboard
df.to_excel('World GDP.xlsx', index = False)
```
